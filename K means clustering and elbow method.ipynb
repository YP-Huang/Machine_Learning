###### 1. Import libraries and create data set with make_blobs

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300, n_features=2, centers=5, cluster_std=2,
          random_state=101)
```

```python
data[0].shape
```

```python
data[1]
```

```python
plt.scatter(data[0][:, 0], data[0][:,1], c = data[1], cmap = 'rainbow')
```

###### 2. k means clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:,1], c = data[1])
```
